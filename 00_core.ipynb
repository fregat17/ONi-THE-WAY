# ONi csv processor

> Read localizations .csv from oxford nanoimager ONi and work with it.

```python
#hide
from nbdev.showdoc import *
```

```python
import pandas as pd
import vaex as v

import os
```

```python
PATH="test_csvs/"

afiles = os.listdir(PATH)

files=[]
for i in afiles:
    if '.csv' in i:
        files.append(i)
        
for i in range(len(files)):
    print('['+str(i)+']'+ files[i])
```

```python
# csvs were saved with unnecessary 'Unnamed: 0' column, read without it.
full = pd.read_csv(PATH + files[0]).drop('Unnamed: 0', axis=1)
long = pd.read_csv(PATH + files[1]).drop('Unnamed: 0', axis=1)
short = pd.read_csv(PATH + files[2]).drop('Unnamed: 0', axis=1)
```

```python
# full column set from ONi localizations export, includes all possible parameters for each localization
print(full.columns)
full
```

```python
# long column set, includes all parameters for localizations except 'Sigma X var' and 'Sigma Y var'
print(long.columns)
long
```

```python
# short column set, includes only Channel, Frame, X/Y/Z in nm, Photons and Background
print(short.columns)
short
```

```python
# need to rename columns with space in names. Vaex can't correctly read data from pandas columns including space in names.
# So we remove all spaces in names (and brackets just in case).
vat = long.rename(columns={'X (nm)':'Xnm', 'Y (nm)':'Ynm', 'Z (nm)':'Znm', 'X precision (nm)':'Xprecisionnm',
       'Y precision (nm)':'Yprecisionnm', 'X (pix)':'Xpix', 'Y (pix)':'Ypix', 'Z (pix)':'Zpix',
       'X precision (pix)':'Xprecisionpix', 'Y precision (pix)':'Yprecisionpix',
       'PSF Sigma X (pix)':'PSFSigmaXpix', 'PSF Sigma Y (pix)':'PSFSigmaYpix'})
```

```python
# read data from pandas dataframe, make vaex fast dataframe
faster = v.from_pandas(vat)
```

```python
faster
```

```python
# widget heatmap that plots our localizations. log1p transform need for correct intensity scaling. 
faster.widget.heatmap(faster['Ypix'], faster['Xpix'], transform='log1p', shape=400)
```
